## Experiment 1: Case Study of Q-Learning on Warehouse Environment <a id='Experiment1'></a> 

In this section we will test our Q_Learning class on the environment

```python
# Define Parameter
N = 10
epsilon = 1
decay = 0.995
alpha = 0.5
gamma = 0.7

# Initiate environment
warehouse = Warehouse(N)

warehouse.reset()
print('Warehouse Environment')
warehouse.display()

# Initiate policy and learning method
e_policy = E_Greedy_Policy(epsilon, decay)
qlearning = Q_Learning(warehouse, alpha, gamma)
```

```python
# Perform Q_Learning
eps_list = []
max_reward_list = []
mean_reward_list = []
variance_list = []
avg_reward_list = []
for eps in range(1000):
    q_learning_episode(qlearning, warehouse,e_policy)
    e_policy.update_epsilon()
    if (eps % 100 == 0) | (eps == 999):
        _, max_rew, mean_rew, var_rew,avg_rew = run_experiments(warehouse, e_policy, qlearning, 100)
        
        #Collect the information
        eps_list.append(eps)
        max_reward_list.append(max_rew)
        mean_reward_list.append(mean_rew)
        variance_list.append(var_rew)
        avg_reward_list.append(avg_rew)
        print('eps:', eps,'max_rew',max_rew ,'mean_reward:', mean_rew, 'epsilon:', round(e_policy.epsilon,6), \
            'variance:', round(var_rew,2), 'Average Reward Per Action:',round(avg_rew,2))

# Save result to df
dict = {'epsiode': eps_list, 'max_reward': max_reward_list, 'mean_reward': mean_reward_list,
            'variance': variance_list,'average_reward_per_action':avg_reward_list}
df_result = pd.DataFrame(dict)
```

```python
#plot the graph
df_temp_1 = pd.melt(df_result, id_vars=['epsiode'], value_vars=['max_reward', 'mean_reward'])

fig, ax =plt.subplots(2, figsize= (12,8))

plot1 = sns.lineplot(x = "epsiode", y = "value",hue = 'variable' ,legend = 'full',data = df_temp_1, ax = ax[0])
plot2 = sns.lineplot(x = "epsiode", y = 'variance',color='orange' , ax= ax[1],data = df_result)
```

The mean_reward was continuously improved from starting point and almost converge with the maximum reward after training for 500 episodes. The variance also decreased to a stable point at around 800 epidosed. This infer theat our implementation of Q-Learning is working correctly.

## Experiment 2: Study the effect of different hyperparameters <a id='Experiment2'></a> 

The parameter in Q Learning consists of starting epsilon, decay, learning rate (alpha) and gamma. <br>
In this seciton we will fix the starting epsilon to be 1. This means that the policy will have a high chance of exploration at the starting of Q learning.
The decay is set to be 0.95 and update the epsilon every epsiode.
This means that the policy allows more time for the agent to explore the environment before taking on the greedy action.
This method of implementation allows the agent to perform well in large environment but with the tradeoff of taking more time to train in the small environment

This section will study the effect of alpha and gamma by setting both parameter to be an extreme value.

# Experiment 3: Evaluate the performance of the algorithm against the environment size <a id='Experiment3'></a> 

In this section, we try to test Q-Learning algorithm against when the Warehouse environment complexity is increasing by increased its size.
The increase in possible states of the environment ususally pose a problem to the algoirthm since it will be struggle for the algorithm to try all action in all states

<b> This step takes long time (around 2 hours for me)

```python
# An additional helper function to automate the Q Learning against many value of environment size
def get_size_result(size_list, alpha, gamma, epsilon, decay, eps):
    '''
    A function that extract the rewards for each environment size when its has trained up to its maximum episode
    
    :param size_list: List of possible size for the environment
    :param alpha: alpha for Q Learning
    :param gamma: gamma for Q Learning
    :param eps: Upper limit of training eps
    :return: result dataframe
    '''
    max_container = []
    mean_container = []
    env_container =[]
    var_container =[]
    avg_container = []
    
    for size in size_list:
        env = Warehouse(size)
        policy = E_Greedy_Policy(epsilon, decay)
        learning_method = Q_Learning(env, alpha, gamma)


        for episode in range(eps):
            q_learning_episode(learning_method, env, policy)
            policy.update_epsilon()

        _, max_rew, mean_rew, var_rew,avg_rew = run_experiments(env, policy, learning_method, 100)

        max_container.append(max_rew)
        mean_container.append(mean_rew)
        var_container.append(var_rew)
        env_container.append(size)
        avg_container.append(avg_rew)
        


    dict = {'size':env_container, 'max_reward': max_container, 'mean_reward': mean_container,
            'variance': var_container,'average_reward':avg_container}
    df_result = pd.DataFrame(dict)

    return df_result
```

```python
# This step takes long time (around 2 hours for me)
size_range = [10,20,30,40,50]

alpha = 0.5
gamma = 0.9
epsilon = 1
decay = 0.995
df_size = get_size_result(size_range, alpha, gamma, epsilon, decay, 5000)
```

```python
df_size.rename({'average_reward': 'average_reward_per_action'}, axis=1, inplace=True)
fig, ax = plt.subplots()

# Plot of an average reward per action against the environment size
sns.lineplot(x = "size", y = "average_reward_per_action",data = df_size,color = 'seagreen')
ax.set_xlim(5,55)
ax.set_xticks([10,20,30,40,50])
```

```python
# Plot of a variance against the environment size
fig, ax = plt.subplots()
sns.lineplot(x = "size", y = "variance",data = df_size,color ='orange')
ax.set_xlim(5,55)
ax.set_xticks([10,20,30,40,50])
```

```python
# Save the result
df_size.to_csv('df_size.csv')
```
